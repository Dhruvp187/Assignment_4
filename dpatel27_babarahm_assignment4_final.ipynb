```python
import numpy as np
import matplotlib.pyplot as plt
import gym
from gym import spaces
import random
from IPython.display import clear_output
from IPython.core.pylabtools import figsize
```

```python
from google.colab import widgets
import time

output_grid = widgets.Grid(1, 1)
for _ in range(1):
  grid = np.zeros((4, 4))
  grid[np.random.randint(4), np.random.randint(4)] = 1
  with output_grid.output_to(0, 0):
    output_grid.clear_cell()
    plt.imshow(grid)
  time.sleep(1)
```

```python
class GridEnvironment(gym.Env):
  metadata = { 'render.modes': [] }

  def __init__(self):
      self.observation_space = spaces.Discrete(16)
      self.action_space = spaces.Discrete(4)
      self.max_timesteps = 15
      self.reward1_done = False
      self.reward2_done = False

  def reset(self):
      self.timestep = 0
      self.agent_pos = [0, 0]
      self.goal_pos = [3, 3]
      self.reward1 = [0, 3]
      self.negative_reward = [2, 2]
      self.negative_reward2 = [0,2]
      self.reward2 = [2,1]
      self.state = np.zeros((4,4))
      self.state[tuple(self.agent_pos)] = 0.9
      self.state[tuple(self.goal_pos)] = 0.8
      self.state[tuple(self.reward1)] = 0.6
      self.state[tuple(self.reward2)] = 0.5
      self.state[tuple(self.negative_reward)] = 0.3
      self.state[tuple(self.negative_reward2)] = 0.2
      observation = self.state.flatten()
      return self.agent_pos

  def step(self, action):
      if action == 0:
        self.agent_pos[0] += 1
      if action == 1:
        self.agent_pos[0] -= 1
      if action == 2:
        self.agent_pos[1] += 1
      if action == 3:
        self.agent_pos[1] -= 1

      self.agent_pos = np.clip(self.agent_pos, 0, 3)
      self.state = np.zeros((4,4))
      self.state[tuple(self.agent_pos)] = 1
      self.state[tuple(self.goal_pos)] = 0.5
      observation = self.state.flatten()

      reward = 0

      if (self.agent_pos == self.reward1).all():
        if(not self.reward1_done):
          reward = 5
          self.reward1_done = True
      if (self.agent_pos == self.negative_reward).all():
        reward = -10
      if (self.agent_pos == self.negative_reward2).all():
        reward = -10
      if (self.agent_pos == self.reward2).all():
        reward =15
        # if(not self.reward2_done):
        #   reward = 15
        #   self.reward2_done = True
      if (self.agent_pos == self.goal_pos).all():
        reward = 25

      self.timestep += 1
      done = True if ((self.timestep >= self.max_timesteps) or (self.agent_pos == self.goal_pos).all()) else False
      info = {}

      return self.agent_pos, reward, done, info

  def render(self):
      self.state[tuple(self.reward1)] = 0.6
      self.state[tuple(self.reward2)] = 0.5
      self.state[tuple(self.negative_reward)] = 0.3
      self.state[tuple(self.negative_reward2)] = 0.2

      self.state[tuple(self.goal_pos)] = 0.8
      self.state[tuple(self.agent_pos)] = 0.9
      plt.imshow(self.state)
      plt.show()
```

```python
env = GridEnvironment()
obs = env.reset()
env.render()
```

```python
class RandomAgent:
  def __init__(self, env):
    self.env = env
    self.observation_space = env.observation_space
    self.action_space = env.action_space

  def step(self, observation):
    return np.random.choice(self.action_space.n)
```

```python
env = GridEnvironment()
agent = RandomAgent(env)

obs = env.reset()
done = False

output_grid = widgets.Grid(1, 1)
with output_grid.output_to(0, 0):
  env.render()

while not done:
  action = agent.step(obs)
  obs, reward, done, info = env.step(action)
  with output_grid.output_to(0, 0):
    output_grid.clear_cell()
    env.render()
  print("Reward:",reward)
  time.sleep(1)
```

#**Part II**

```python
env = GridEnvironment()
epsilon = 0.9
learning_rate = 0.2
gamma = 0.8
Sepsilons = []
Srewards = []
Qtable = np.zeros((4,4,env.action_space.n))
observation = env.observation_space
```

```python
episodes = 201
output_grid = widgets.Grid(1, 1)

for episode in range(episodes):
  obs = env.reset()
  done = False
  Sepsilons.append(epsilon)
  current_reward = 0
  while done == False:
    if(episode > 195):
      with output_grid.output_to(0, 0):
        output_grid.clear_cell()
        env.render()
      time.sleep(0.1)
    if np.random.uniform(0,1) < epsilon:
     action = np.random.choice(env.action_space.n)
    else:
     action = np.argmax(Qtable[obs[0],obs[1]])

    state = np.copy(obs)
    obs, reward, done, info = env.step(action)
    current_reward = current_reward + reward
    new_state = np.copy(obs)

    prev = Qtable[state[0]][state[1]][action]
    td = reward + (gamma*Qtable[new_state[0]][new_state[1]][action]) - prev
    new = prev + learning_rate*td
    Qtable[state[0]][state[1]][action] = new
  Srewards.append(current_reward)
  epsilon = max(0.1,epsilon * 0.988)
```

#Graphs

```python
plt.figure(figsize=(20,10))
plt.xlabel('Sarsa Deterministic Episode')
plt.ylabel('Epsilon')
plt.plot(Sepsilons)
```
